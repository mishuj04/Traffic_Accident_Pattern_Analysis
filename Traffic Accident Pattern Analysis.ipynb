## Handling Missing Values

```python
df1.isnull().sum()
```

```python
df1["Driver License Status"].unique()
```

```python
df1["Traffic Control Presence"].unique()
```

```python
df1["Driver License Status"].fillna(df1["Driver License Status"].mode()[0], inplace= True)
```

```python
df1["Traffic Control Presence"].fillna(df1["Traffic Control Presence"].mode()[0], inplace= True)
```

```python
df1.isnull().sum()
```

```python
df1.head()
```

```python
df1["Month"]=df1["Month"].map({"January":1, "February":2, "March":3, "April":4, "May":5, "June":6, "July":7, "August":8, "September":9, "October":10, "November":11, "December":12})
```

```python
df1.head()
```

```python
df2=df1.copy()
```

```python
df2.head()
```

```python
df2["Driver License Status"].unique()
```

```python
df2["Driver License Status"]=df2["Driver License Status"].map({"Valid":1, "Expired":0})
```

```python
df2.head()
```

```python
df2["Alcohol Involvement"].unique()
```

```python
df2["Alcohol Involvement"]=df2["Alcohol Involvement"].map({"Yes":1, "No":0})
```

```python
df2.head()
```

```python
df2["Hour"] = pd.to_datetime(df2["Time of Day"], format="%H:%M").dt.hour
df2["Minute"] = pd.to_datetime(df2["Time of Day"], format="%H:%M").dt.minute
```

```python
df2.head()
```

```python
def get_time_period(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

df2["Time Period"] = df2["Hour"].apply(get_time_period)
```

# Key Findings from Traffic Accident Analysis

### Temporal Patterns
##### Seasonal Impact: October recorded the highest number of fatalities, while July had the lowest
##### Annual Trends: 2021 emerged as the year with the most fatalities
##### Weekly Patterns: Wednesday showed the highest accident severity rates
##### Time of Day: Nighttime accidents resulted in over 2,500 fatalities, significantly exceeding daytime incidents

### Demographic Insights
##### Age Distribution: The 45-59 age group experienced the highest fatality rates, while the 0-17 age group had the lowest
##### Gender Patterns: Female drivers were involved in more accidents than male drivers
##### Alcohol-Related Incidents: Among drunk driving cases, female drivers had higher accident rates than male drivers

### Vehicle and Infrastructure Analysis
##### Vehicle Type Safety: Cars had the lowest fatality rates compared to other vehicle types
##### Road Infrastructure: State Highways accounted for the majority of accidents relative to other road types

### Geographic Distribution
##### Regional Variations: Goa reported the highest number of casualties, while Arunachal Pradesh recorded the lowest

# Conclusions

### Immediate Actions Required
##### Infrastructure: Upgrade State Highway safety and nighttime lighting
##### Enforcement: Increase Wednesday traffic monitoring and night patrol
##### Campaigns: Target October safety drives and middle-aged driver education

### Key Risk Factors Identified
##### Time: October months and nighttime hours are most dangerous
##### Demographics: Ages 45-59 and female drivers need focused attention
##### Location: State Highways and high-traffic states require priority intervention

### Policy Focus Areas
##### Vehicle Safety: Strengthen standards for non-car vehicles
##### Alcohol Prevention: Gender-specific drunk driving programs
##### Resource Allocation: Redistribute safety resources based on regional casualty data
